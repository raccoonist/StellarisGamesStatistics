# Stellaris Game 3 Statistics

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize']=(10,10)
```

```python
# file paths
populationPath = "data/population.csv"
techScorePath = "data/techscore.csv"
economyScorePath = "data/economyscore.csv"
```

```python
# read in all csvs
population = pd.read_csv(populationPath)
techScore = pd.read_csv(techScorePath)
economyScore = pd.read_csv(economyScorePath)
```

```python
ax = sns.lineplot(x='Year', y='value',hue='variable',data=pd.melt(population, ['Year']))
ax.set(ylabel='Population', title='Population Over Time')
legend = ax.legend()
legend.set(title="Players")
plt.grid()
plt.show()
```

```python
ax = sns.lineplot(x='Year', y='value',hue='variable',data=pd.melt(techScore, ['Year']))
ax.set(ylabel='Tech Score', title='Tech Score Over Time')
legend = ax.legend()
legend.set(title="Players")
plt.grid()
plt.show()
```

```python
ax = sns.lineplot(x='Year', y='value',hue='variable',data=pd.melt(economyScore, ['Year']))
ax.set(ylabel='Economy Score', title='Economy Score Over Time')
legend = ax.legend()
legend.set(title="Players")
plt.grid()
plt.show()
```

```python
common_years = np.intersect1d(np.intersect1d(population.loc[:,'Year'].values, techScore.loc[:,'Year'].values),economyScore.loc[:,'Year'].values)
print(common_years)
```

```python
populationScore = 2*population
populationScore['Year'] = population['Year']

leaderboardScore = pd.concat([populationScore, techScore, economyScore]).groupby(by=['Year']).sum().reset_index()
leaderboardScore = leaderboardScore[leaderboardScore['Year'].isin(common_years)].reset_index(drop=True)
leaderboardScore
```

```python
ax = sns.lineplot(x='Year', y='value',hue='variable',data=pd.melt(leaderboardScore, ['Year']))
ax.set(ylabel='Total Score')
legend = ax.legend()
legend.set(title="Players")
plt.grid()
plt.show()
```
