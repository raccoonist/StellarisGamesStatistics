# Stellaris Game 3 Statistics

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize']=(10,10)
```

```python
# file paths
populationPath = "data/population.csv"
techScorePath = "data/techScore.csv"
economyScorePath = "data/economyScore.csv"
```

```python
# read in all csvs
population = pd.read_csv(populationPath)
techScore = pd.read_csv(techScorePath)
economyScore = pd.read_csv(economyScorePath)
```

```python
ax = sns.lineplot(x='Year', y='value',hue='variable',data=pd.melt(population, ['Year']))
ax.set(ylabel='Population')
legend = ax.legend()
legend.set(title="Players")
plt.grid()
plt.show()
```

```python
ax = sns.lineplot(x='Year', y='value',hue='variable',data=pd.melt(techScore, ['Year']))
ax.set(ylabel='Tech Score')
legend = ax.legend()
legend.set(title="Players")
plt.grid()
plt.show()
```

```python
ax = sns.lineplot(x='Year', y='value',hue='variable',data=pd.melt(economyScore, ['Year']))
ax.set(ylabel='Economy Score')
legend = ax.legend()
legend.set(title="Players")
plt.grid()
plt.show()
```

```python
common_years = np.intersect1d(population.loc[:,'Year'].values, techScore.loc[:,'Year'].values)
print(common_years)
```

```python
leaderboardScore = pd.DataFrame(common_years, columns=['Year'])
leaderboardScore = leaderboardScore.merge(population, how='left')
leaderboardScore
```
