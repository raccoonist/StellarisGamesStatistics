# Stellaris Game 3 Statistics

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = 10,10
```

```python
# file paths
population_path = "data/population.csv"
tech_score_path = "data/techscore.csv"
economy_score_path = "data/economyscore.csv"
```

```python
# read in all csvs
population = pd.read_csv(population_path)
tech_score = pd.read_csv(tech_score_path)
economy_score = pd.read_csv(economy_score_path)
```

## Main Statistics

```python
population
```

```python
tech_score
```

```python
economy_score
```

```python
ax = sns.lineplot(x='Year', y='value',hue='variable',data=pd.melt(population, ['Year']))
ax.set(ylabel='Population', title='Population Over Time')
legend = ax.legend()
legend.set(title="Players")
plt.grid()
plt.show()
```

```python
ax = sns.lineplot(x='Year', y='value',hue='variable',data=pd.melt(tech_score, ['Year']))
ax.set(ylabel='Tech Score', title='Tech Score Over Time')
legend = ax.legend()
legend.set(title="Players")
plt.grid()
plt.show()
```

```python
ax = sns.lineplot(x='Year', y='value',hue='variable',data=pd.melt(economy_score, ['Year']))
ax.set(ylabel='Economy Score', title='Economy Score Over Time')
legend = ax.legend()
legend.set(title="Players")
plt.grid()
plt.show()
```

```python
common_years = np.intersect1d(np.intersect1d(population.loc[:,'Year'].values, tech_score.loc[:,'Year'].values), economy_score.loc[:,'Year'].values)
print(common_years)
```

```python
population_score = 2*population
population_score['Year'] = population['Year']

leaderboard_score = pd.concat([population_score, tech_score, economy_score]).groupby(by=['Year']).sum().reset_index()
leaderboard_score = leaderboard_score[leaderboard_score['Year'].isin(common_years)].reset_index(drop=True)
leaderboard_score
```

```python
ax = sns.lineplot(x='Year', y='value',hue='variable',data=pd.melt(leaderboard_score, ['Year']))
ax.set(ylabel='Total Score', title='Total Score Over Time')
legend = ax.legend()
legend.set(title="Players")
plt.grid()
plt.show()
```
